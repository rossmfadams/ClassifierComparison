#Logistic Regression



```
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```
iris = datasets.load_iris()
print(iris)
X = iris.data
y = iris.target

numclasses = len(iris.target_names)

# Split the data into 50% training and 50% testing
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.5,\
                                                  random_state = 5, stratify=y)

# Train the model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```
# Test the model
predictLR = model.predict(X_test)
print(predictLR)

print() # New line

# Check precision, recall, f1-score
print(classification_report(y_test, predictLR))

print("Accuracy Score: ",accuracy_score(y_test, predictLR))
cnf_matrix = confusion_matrix(y_test, predictLR)
print('confusion matrix = ')
print(cnf_matrix)
```

#K-Nearest Neighbor

```
# Load the iris database
iris = datasets.load_iris()
X = iris.data
Y = iris.target

# Use the KNN model on the database split
modelKNN = KNeighborsClassifier(n_neighbors = 1, metric = 'euclidean')
modelKNN.fit(X_train, y_train)

# Collect predictions
predictKNN = modelKNN.predict(X_test)
print(predictKNN)

# Print the report
print(classification_report(y_test, predictKNN))

# Print the confusion matrix
modelKNN_matrix = confusion_matrix(y_test, predictKNN)
print('confusion matrix = ')
print(modelKNN_matrix)
```

#Binary Decision Tree

```
# Load the iris database
iris = datasets.load_iris()
X = iris.data
Y = iris.target

# Use the decisiontree on the database
modelDT = DecisionTreeClassifier()
modelDT.fit(X_train, y_train)

# Collect predictions
predictDT = modelDT.predict(X_test)
print(predictDT)

# Print the report
print(classification_report(y_test, predictDT))

# Print the confusion matrix
modelDT_matrix = confusion_matrix(y_test, predictDT)
print('confusion matrix = ')
print(modelDT_matrix)

# Plot the decision tree as a figure
fig = plt.figure(figsize = (25,20))
tree_diagramPlot = tree.plot_tree(modelDT, feature_names = iris.feature_names, class_names = iris.target_names, filled = True)
```
